```python
order_date_col = 'Order Date'
ship_date_col = 'Ship Date'
sales_col = 'Total Revenue'
```

# Data Cleaning & Droping 

```python
df_cleaned = df.dropna(subset = [order_date_col,ship_date_col,sales_col])
```

```python
df_cleaned[order_date_col] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df_cleaned[ship_date_col] = pd.to_datetime(df['Ship Date'], errors='coerce')
```

```python
df_cleaned.info()
```

```python
df.isnull().sum()
```

```python
# Extract Year , Month and Month-Year from Order Date

df_cleaned['Year'] = df_cleaned[order_date_col].dt.year
df_cleaned['Month'] = df_cleaned[ship_date_col].dt.strftime('%B')
df_cleaned['MonthYear'] = df_cleaned[order_date_col].dt.strftime('%b %Y')
```

```python
print(df_cleaned.head())
```

```python
date_table = pd.DataFrame({
    'Date': pd.date_range(start=df_cleaned[order_date_col].min(), end=df_cleaned[ship_date_col].max())
})
date_table['Year'] = date_table['Date'].dt.year
date_table['Month'] = date_table['Date'].dt.strftime('%B')
date_table['MonthYear'] = date_table['Date'].dt.strftime('%b %Y')
```
